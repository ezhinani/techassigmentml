```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
```

```python
df = pd.read_csv('ai4i2020.csv')
```

```python
#Untuk melihat 5 baris teratas dari sebuah data.
df.head()
```

```python
#Untuk mengetahui jumlah baris dan kolom dalam tabel data.
df.shape
```

```python

#Untuk menampilkan data ringkas dalam DataFrame.
df.info()
```

```python

#Untuk memberikan ringkasan statistik dari DataFrame untuk kolom numerik secara default.
df.describe()
```

```python

#Digunakan untuk menghitung jumlah nilai null di setiap kolom dalam DataFrame.
df.isnull().sum()
```

```python

#Digunakan untuk mengidentifikasi baris duplikat dalam DataFrame.
df.duplicated()
```

```python

#Digunakan untuk menghitung jumlah baris duplikat dalam DataFrame.
df.duplicated().sum()
```

```python

#Digunakan untuk menghapus kolom-kolom yang disebutkan dari DataFrame
df.drop(['UDI','Product ID','TWF'], axis=1, inplace=True)
```

```python

#Digunakan untuk menghapus kolom-kolom yang disebutkan dari DataFrame
df.drop(['Type', 'HDF', 'PWF', 'OSF', 'RNF'], axis=1, inplace=True)
```

```python

#Untuk melihat 5 baris teratas dari sebuah data.
df.head()
```

```python
Input_columns = ['Air temperature [K]' , 'Process temperature [K]', 'Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]']
```

```python
Output_columns = ['Machine failure']
```

```python
X = df[Input_columns]
y = df[Output_columns]
```

```python

#Untuk menampilkan 5 data teratas.
X.head()
```

```python

#Untuk menampilkan 5 data teratas.
y.head()
```

```python

#Digunakan untuk membagi dataset menjadi data pelatihan dan data pengujian.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python

#Perintah untuk membuat sebuah instance dari kelas LinearRegression.
lin_reg = LinearRegression()
```

```python

#Digunakan untuk melatih (fit) model regresi linier pada data pelatihan.
lin_reg.fit(X_train, y_train)
```

```python

#Digunakan untuk menampilkan koefisien dan intersep (intercept) dari model regresi linier.
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python

#Digunakan untuk mengevaluasi kinerja model regresi linier pada data pengujian.
lin_reg.score(X_test, y_test)
```

```python

#Digunakan untuk membuat dan melatih model decision tree untuk klasifikasi.
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
```

```python

#Digunakan untuk membuat prediksi kelas target berdasarkan fitur-fitur dalam data pengujian menggunakan model pohon keputusan yang telah dilatih.
y_pred = clf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Akurasi: {accuracy}')
print('Classification Report:')
print(classification_report(y_test, y_pred))
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))
```

```python
#Digunakan untuk evaluasi model Decission Tree nilai precision 80% recall 80% acc78%
y_pred = clf.predict(X_test)
```

```python
#Digunakan untuk mengevaluasi kinerja model.
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
```

```python
print(f'Akurasi: {accuracy:.2f}')
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print('Classification Report:')
print(classification_report(y_test, y_pred))
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))
```

```python
#Digunakan untuk evaluasi Decision Tree.
accuracy_dt = accuracy_score(y_test, y_pred_dt)
precision_dt = precision_score(y_test, y_pred_dt, average='weighted')
recall_dt = recall_score(y_test, y_pred_dt, average='weighted')

#Digunakan untuk evaluasi Logistic Regression.
accuracy_lr = accuracy_score(y_test, y_pred_lr)
precision_lr = precision_score(y_test, y_pred_lr, average='weighted')
recall_lr = recall_score(y_test, y_pred_lr, average='weighted')
```

```python
print("Decision Tree Evaluation")
print(f'Akurasi: {accuracy_dt:.2f}')
print(f'Precision: {precision_dt:.2f}')
print(f'Recall: {recall_dt:.2f}')
print('Classification Report:')
print(classification_report(y_test, y_pred_dt))
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred_dt))

print("\nLogistic Regression Evaluation")
print(f'Akurasi: {accuracy_lr:.2f}')
print(f'Precision: {precision_lr:.2f}')
print(f'Recall: {recall_lr:.2f}')
print('Classification Report:')
print(classification_report(y_test, y_pred_lr))
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred_lr))
```
